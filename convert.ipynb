```
#%cd ..
%cd Darknetproject
import pandas as pd
import re

column_name = ["filename","object","top","left","height","width"]
 
    
inputdf = pd.read_csv("export-2020-05-08T01_20_04.078Z.csv")
index = list(range(0, inputdf.shape[0])) 
outdf = pd.DataFrame(index=index, columns=column_name)

#print(inputdf.shape)
#print(outdf.shape)

for i in range(inputdf.shape[0]):  
  outdf["filename"][i]=re.search(r'(?<=dog.).*?(?=jpg)', inputdf["External ID"][i]).group(0)
  outdf["object"][i]=re.search(r'(?<="title":").*?(?=",)', inputdf["Label"][i]).group(0)
  outdf["top"][i]=re.search(r'(?<="top":).*?(?=,)', inputdf["Label"][i]).group(0)
  outdf["left"][i]=re.search(r'(?<="left":).*?(?=,)', inputdf["Label"][i]).group(0)
  outdf["height"][i]=re.search(r'(?<="height":).*?(?=,)', inputdf["Label"][i]).group(0)
  outdf["width"][i]=re.search(r'(?<="width":).*?(?=})', inputdf["Label"][i]).group(0)
 
 
new_row = {"filename":"383." ,'object':'dog', 'top':"87", 'left':"92", 'height':"97", "width":"132"}
 
 
 
outdf = outdf.append(new_row, ignore_index=True)
 
```
